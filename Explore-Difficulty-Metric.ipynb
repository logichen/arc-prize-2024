# Explore a Difficulty Metric

This notebook is an experimental approach on solving the ARC challenge. See https://arcprize.org/ for more details.

```python
import random
import torch

from pathlib import Path

from arc2024.data import FewShotDataset
from arc2024.utils import plot, calc
```

## Prepare Dataset

```python
dataset = FewShotDataset(dataset_dir=Path('datasets'))
dataset
```

## Rate Difficulty of each Challenge from Training Dataset

```python
X, y, challenge_id = random.choice(dataset)
X_samples, y_samples = dataset.get_samples(challenge_id)

print("Examples: (Difficulty = " + str(calc.hamming_distance(X_samples, y_samples)) + ")")
for i, X_sample in enumerate(X_samples):
    # print("Difficulty = " + str(calc.hamming_distance(X_sample, y_samples[i])))
    plot.input_and_output(
        X_sample,
        y_samples[i]
    )

print("Task: (Difficulty = " + str(calc.hamming_distance(X, y)) + ")")
plot.input_and_output(X, y)

# print(f"y.shape: {y.shape}")
# fft = torch.fft.fft(y)
# plot.as_is(torch.view_as_real(fft))
```
